```python
import pandas as pd
import numpy as np
import seaborn as sns
sns.set_style('whitegrid')
```

```python
std_dev = 10
num_simulations = 100000
```

```python
data = pd.read_excel('C:\\Users\\kchandrakant\\Downloads\\Problem_set_2_02.xlsx')
```

```python
discount_rate = data.iloc[1,1]
```

```python
initial_investment = np.absolute(data.iloc[0,1])
```

```python
avg_cash_flow = data.iloc[0,2:].to_numpy(dtype='float64')
```

```python
time_period = np.zeros(avg_cash_flow.size)
for i in range(avg_cash_flow.size):
    time_period[i]=i
```

```python
nav_array = np.zeros(num_simulations)
```

```python
for i in range(num_simulations):
    random_cash_flow = np.random.normal(avg_cash_flow, std_dev).round(2)
    pv_array = (random_cash_flow/(np.power((1+discount_rate), time_period)))
    nav = (pv_array.sum()-initial_investment).round(2)
    nav_array[i]=nav
```

```python
sns.distplot(nav_array)
```
