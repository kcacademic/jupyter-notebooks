```python
env = gym.make('Acrobot-v1')
print(env.action_space) #[Output: ] Discrete(3)
print(env.observation_space) #[Output: ] Box(6,)
env = gym.make('CartPole-v1')
print(env.action_space) #[Output: ] Discrete(2)
print(env.observation_space) # [Output: ] Box(4,)
env = gym.make('MountainCar-v0')
print(env.action_space) #[Output: ] Discrete(3)
print(env.observation_space) #[Output: ] Box(2,)
env = gym.make('MountainCarContinuous-v0')
print(env.action_space) #[Output: ] Box(1,)
print(env.observation_space) #[Output: ] Box(2,)
env = gym.make('Pendulum-v0')
print(env.action_space) #[Output: ] Box(1,)
print(env.observation_space) #[Output: ] Box(3,)

env = gym.make('Blackjack-v0')
print(env.action_space) #[Output: ] Discrete(2)
print(env.observation_space) #[Output: ] Tuple(Discrete(32), Discrete(11), Discrete(2))
env = gym.make('FrozenLake-v0')
print(env.action_space) #[Output: ] Discrete(4)
print(env.observation_space) #[Output: ] Discrete(16)
env = gym.make('FrozenLake8x8-v0')
print(env.action_space) #[Output: ] Discrete(4)
print(env.observation_space) #[Output: ] Discrete(64)
env = gym.make('GuessingGame-v0')
print(env.action_space) #[Output: ] Box(1,)
print(env.observation_space) #[Output: ] Discrete(4)
env = gym.make('HotterColder-v0')
print(env.action_space) #[Output: ] Box(1,)
print(env.observation_space) #[Output: ] Discrete(4)
env = gym.make('NChain-v0')
print(env.action_space) #[Output: ] Discrete(2)
print(env.observation_space) #[Output: ] Discrete(5)
env = gym.make('Roulette-v0')
print(env.action_space) #[Output: ] Discrete(38)
print(env.observation_space) #[Output: ] Discrete(1)
env = gym.make('Taxi-v3')
print(env.action_space) #[Output: ] Discrete(6)
print(env.observation_space) #[Output: ] Discrete(500)
```

```python
env = gym.make('NChain-v0') # try for different environements
observation = env.reset()
for t in range(100):
        #env.render()
        print (observation)
        action = env.action_space.sample()
        observation, reward, done, info = env.step(action)
        print (observation, reward, done, info)
        if done:
            print("Finished after {} timesteps".format(t+1))
            break
```
